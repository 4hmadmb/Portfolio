```python
from bs4 import BeautifulSoup
import requests
```

```python
url= 'https://www.espn.com/soccer/team/stats/_/id/83/league/ESP.1/season/2022/view/scoring'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

Scraping the Goals Table

```python
title = soup.find_all('th')[:4]
```

```python
title
```

```python
table_title = [title.text.strip() for title in title]
print(table_title)
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(columns = table_title)
df1
```

```python
column_data = soup.find_all('tr')
```

```python
for row in column_data[1:25]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]

  length = len(df1)
  df1.loc[length] = individual_row_data
```

```python
df1
```

Scraping the Assists Table

```python
title_ = soup.find_all('th')[4:]
```

```python
title_
```

```python
table_title_ = [title.text.strip() for title in title_]
print(table_title_)
```

```python
df_ = pd.DataFrame(columns = table_title_)
df_
```

```python
for row in column_data[26:]:
  arow_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in arow_data]

  length = len(df_)
  df_.loc[length] = individual_row_data
```

```python
df_
```

Scraping the Discipline Table

```python
url= 'https://www.espn.com/soccer/team/stats/_/id/83/league/ESP.1/season/2022/view/discipline'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
ti = soup.find_all('th')
```

```python
ti
```

```python
table_ti = [title.text.strip() for title in ti]
print(table_ti)
```

```python
dff = pd.DataFrame(columns = table_ti)
dff
```

```python
columndata = soup.find_all('tr')
```

```python
for row in columndata[1:]:
  drow_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in drow_data]


  length = len(dff)
  dff.loc[length] = individual_row_data
```

```python
dff
```

Merging all three tables together

```python
r = pd.merge(df1, df_, on=['Name','P'], how="outer")
```

```python
r
```

```python
DataF = pd.merge(r, dff, on=['Name','P'], how="outer")
```

```python
DataF
```

Dropping unnecessary columns

```python
df2 = DataF.drop(columns = ['RK_x','RK_y','RK','Pts'])
```

```python
df2
```

```python
df2.rename(columns = {"Name": "Player_Name", "P": "Games_Played", "G": "Goals", "A": "Assists",
             "yc": "Yellow_Card", "rc": "Red_Card"}, inplace = True)
```

```python
df2
```

Fill all Na with 0

```python
df2 = df2.fillna(0)
df2
```
